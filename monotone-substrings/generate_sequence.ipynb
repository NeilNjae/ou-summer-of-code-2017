```python
import sys
import random
```

```python
def generate_sequence(length, smallest=0, biggest=sys.maxsize, same_chance=0.05, flip_chance=0.3):
    seq = [random.randint(smallest, biggest)]
    increasing = random.choice([True, False])
    for _ in range(length):
        last = seq[-1]
        if random.random() < flip_chance:
            increasing = not increasing
        if random.random() < same_chance:
            seq += [last]
        elif increasing:
            seq += [random.randint(last, biggest)]
        else:
            seq += [random.randint(smallest, last)]
    return seq
```

```python
generate_sequence(1000)
```

```python
for i in range(10):
    seq = generate_sequence(10000)
    with open('sequence{:03}.txt'.format(i), 'w') as f:
        f.write(','.join(str(n) for n in seq))
```
